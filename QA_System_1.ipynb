```python
from youtube_transcript_api import YouTubeTranscriptApi
import re

# Function to extract video ID from YouTube URL
def extract_video_id(url):
    match = re.search(r"(?:v=|youtu\.be/)([a-zA-Z0-9_-]{11})", url)
    return match.group(1) if match else None

# Function to fetch and return full transcript text (with line breaks every 10 words)
def get_formatted_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        full_text = " ".join([entry['text'] for entry in transcript])

        # Split into chunks of 10 words
        words = full_text.split()
        lines = [" ".join(words[i:i+10]) for i in range(0, len(words), 10)]

        return "\n".join(lines)
    except Exception as e:
        return f"Error fetching transcript: {e}"

# Main
if __name__ == "__main__":
    youtube_url = input("Enter YouTube video URL: ")
    video_id = extract_video_id(youtube_url)

    if video_id:
        transcript = get_formatted_transcript(video_id)
        print("\n--- Formatted Transcript ---\n")
        print(transcript)
    else:
        print("Invalid YouTube URL. Please check and try again.")
```

```python
from transformers import pipeline

# Summarize the transcript and format output with newlines after every 7 words
def summarize_transcript(context):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

    # Break into manageable chunks
    chunk_size = 2000
    chunks = [context[i:i + chunk_size] for i in range(0, len(context), chunk_size)]

    summaries = []
    for chunk in chunks:
        summary = summarizer(chunk, max_length=130, min_length=30, do_sample=False)
        summaries.append(summary[0]['summary_text'])

    combined_summary = " ".join(summaries)

    # Add newline after every 7 words
    words = combined_summary.split()
    formatted_summary = "\n".join([" ".join(words[i:i+7]) for i in range(0, len(words), 7)])

    return formatted_summary

# Assuming `transcript_text` is already defined
summarized_context = summarize_transcript(transcript)
print("Summarized Transcript:\n", summarized_context)
```

```python
import speech_recognition as sr
from pydub import AudioSegment
from google.colab import files

def convert_to_wav(input_path, output_path="converted.wav"):
    try:
        audio = AudioSegment.from_file(input_path)
        audio.export(output_path, format="wav")
        return output_path
    except Exception as e:
        print(f"Audio conversion failed: {e}")
        return None

def speech_to_text_from_file(file_path):
    recognizer = sr.Recognizer()

    # Convert to .wav if not already
    if not file_path.endswith(".wav"):
        print("Converting audio to WAV format...")
        file_path = convert_to_wav(file_path)
        if not file_path:
            return None

    # Load and process audio
    try:
        with sr.AudioFile(file_path) as source:
            print("Processing audio file...")
            audio = recognizer.record(source)
    except Exception as e:
        print(f"Error loading audio file: {e}")
        return None

    # Speech recognition
    try:
        question = recognizer.recognize_google(audio)
        print(f"Extracted Question: {question}")
        return question
    except sr.UnknownValueError:
        print("Audio unclear. Could not extract text.")
    except sr.RequestError as e:
        print(f"Speech Recognition error: {e}")
    return None

# Upload the audio file
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
question = speech_to_text_from_file(file_path)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the T5 model and tokenizer
t5_model = T5ForConditionalGeneration.from_pretrained("t5-base")
t5_tokenizer = T5Tokenizer.from_pretrained("t5-base")

def answer_question(question, context):
    input_text = f"question: {question} context: {context}"
    inputs = t5_tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)

    outputs = t5_model.generate(inputs, max_length=50, num_beams=4, early_stopping=True)
    answer = t5_tokenizer.decode(outputs[0], skip_special_tokens=True)

    return answer

# Ensure both `question` and `summarized_context` are already available from previous cells
if question and summarized_context:
    answer = answer_question(question, summarized_context)
    print(f"🧠 Answer: {answer}")
else:
    print("Missing question or context.")
```

```python
from transformers import pipeline
import re

# Always use CPU (safe for Colab or any unstable CUDA)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=-1)

def clean_text(text):
    # Optional: Clean transcript (remove multiple spaces, newlines, etc.)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def chunk_text(text, max_tokens=800):
    sentences = text.split('. ')
    chunks, current = [], ""

    for sentence in sentences:
        if len(current.split()) + len(sentence.split()) <= max_tokens:
            current += sentence + ". "
        else:
            if len(current.split()) >= 20:  # Only keep meaningful chunks
                chunks.append(current.strip())
            current = sentence + ". "

    if len(current.split()) >= 20:
        chunks.append(current.strip())

    return chunks

def transcript_to_notes(transcript):
    transcript = clean_text(transcript)
    chunks = chunk_text(transcript)
    notes = []

    for chunk in chunks:
        try:
            summary = summarizer(chunk, max_length=150, min_length=30, do_sample=False)
            notes.append(summary[0]['summary_text'])
        except Exception as e:
            print(f"Skipping a chunk due to error: {e}")

    if not notes:
        print("⚠️ No notes could be generated. Check transcript content.")
        return

    full_summary = " ".join(notes)
    bullet_points = full_summary.split('. ')
    bullet_points = [f"• {point.strip()}." for point in bullet_points if len(point.strip().split()) >= 3]

    print("📝 Key Notes from Transcript:\n")
    for point in bullet_points:
        print(point)

# Run it
transcript_to_notes(transcript)
```

```python
from googlesearch import search

def find_related_websites(transcript, num_results=5):
    # Use first 15 words of transcript as search query
    query = " ".join(transcript.strip().split()[:15])

    print(f"🔍 Searching for similar content related to:\n\"{query}\"\n")
    print("🌐 Top Related Websites:\n")

    for i, url in enumerate(search(query, num_results=num_results), 1):
        if "youtube.com" not in url.lower():  # Skip YouTube links
            print(f"{i}. {url}")

# 🧪 Example usage
find_related_websites(transcript)
```

```python

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
import spacy
from collections import Counter
import matplotlib.pyplot as plt

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Ensure transcript is defined
text = transcript  # 👈 Your transcript text

# --- Tokenization ---
sentences = sent_tokenize(text)
words = word_tokenize(text)

# --- Remove Stopwords ---
stop_words = set(stopwords.words("english"))
filtered_words = [w.lower() for w in words if w.lower() not in stop_words and w.isalpha()]

# --- Lemmatization ---
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# --- POS Tagging ---
pos_tags = nltk.pos_tag(lemmatized_words)
nouns = [word for word, tag in pos_tags if tag.startswith("NN")]

# --- Named Entity Recognition ---
doc = nlp(text)
entities = [(ent.text, ent.label_) for ent in doc.ents]

# --- Sentiment Analysis ---
blob = TextBlob(text)
sentiment = blob.sentiment

# --- Frequency Plot ---
word_freq = Counter(lemmatized_words).most_common(10)
words_plot, counts = zip(*word_freq)
plt.figure(figsize=(8,4))
plt.bar(words_plot, counts, color='orange')
plt.title("Top 10 Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# --- Results ---
print("🧠 Total Sentences:", len(sentences))
print("🔤 Total Words:", len(words))
print("🧹 Filtered Words:", len(filtered_words))
print("\n📌 Top Nouns:", nouns[:10])
print("\n🏷️ Named Entities:", entities[:10])
print(f"\n💬 Sentiment Polarity: {sentiment.polarity:.2f}")
print(f"📈 Sentiment Subjectivity: {sentiment.subjectivity:.2f}")
```
